```python
import numpy as np
import matplotlib.pyplot as plt
#Question - The coefficient 'w' of a polynomial is an array of numbers to be multiplied to 1 comma space x comma space x squared comma space x cubed comma...
#For example, w=[3, 2, 1] means that the polynomial is 3 plus 2 x plus x squared
#Implement a function 'draw_poly(w, start, end)' that draws a polynomial graph given an array of coefficients 'w' and input range 'start', 'end'.
#Check the figure below for example runs.
#Assume that 'w' is not empty and necessary modules have been imported.
#draw_poly([2,1],-2,4) = y= 2+x

import numpy as np
import matplotlib.pyplot as plt
def draw_poly(w, start, end):
    xvals = np.arange(start, end, 0.1)
    y = w[0]
    for i in range(1,len(w)):
        y += w[i] * xvals**i
    plt.plot(xvals, y)
    plt.show()
draw_poly([4,5,6],1,5)
```
