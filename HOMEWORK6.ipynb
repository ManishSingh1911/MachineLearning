```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats 
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
```

```python
# This is an example to show how to use subplots, setting figure size, setting titles, and scaling.
np.random.seed(0)
x1 = np.random.normal(0,1,100)
y1 = np.random.normal(0,1,100)
figure, ax = plt.subplots(1,3)  # 1 row, 3 column
figure.set_size_inches(10,5)    # width:10 inches, height:5 inches
ax[0].scatter(x1,y1)
ax[0].set_title("alpha=1")
ax[0].axis('scaled')
ax[1].scatter(x1,y1,alpha=0.5)
ax[1].set_title("alpha=0.5")
ax[1].axis('scaled')
ax[2].scatter(x1,y1,alpha=0.2)
ax[2].set_title("alpha=0.2")
ax[2].axis('scaled')
plt.show()
```

```python
# Generate the following data points randomly
# 100 data points from normal distribution with mean=(0,0), sd=(1,1)
# 100 data points from normal distribution with mean=(4,0), sd=(1,4)
np.random.seed(0)
figure, ax = plt.subplots(1,3)
figure.set_size_inches(10,5)

#ADD YOUR CODE HERE 

x1= np.random.normal(0,1,100)
y1= np.random.normal(0,1,100)
ax[0].set_title("Original")
ax[0].scatter(x1,y1,alpha=0.5)

x2= np.random.normal(4,1,100)
y2= np.random.normal(0,4,100)
ax[0].scatter(x2,y2,alpha=0.5)
ax[0].axis('scaled')

# Show the generated data points with each group in different color.
# Using scikit-learn module,
# Run Kmeans algorithm on the data points to show two clusters in different colors.
# Run Gaussina Mixture EM algorithm on the data points to show two clusters in different colors.

#Kmeans algorithm
x= np.hstack([x1,x2])
y= np.hstack([y1,y2])
data= np.vstack([x,y]).T

kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
C=kmeans.labels_ 
ax[1].set_title("Kmeans")
ax[1].scatter(x,y,c=C,alpha=0.5)
ax[1].axis('scaled')

#Gaussian Mixture - EM Algorithm 
EM= GaussianMixture(n_components=2)
EM.fit(data)
cluster=EM.predict(data)
ax[2].set_title("GaussianMixture")
ax[2].scatter(x,y,c=cluster,alpha=0.5)
ax[2].axis('scaled')
```

```python
# Generate the following data points randomly
# 500 data points from normal distribution with mean=(0,0), sd=(2,2)
# 100 data points from normal distribution with mean=(-3,4), sd=(0.5,0.5)
# 100 data points from normal distribution with mean=(3,4), sd=(0.5,0.5)

np.random.seed(0)
figure, ax = plt.subplots(1,3)
figure.set_size_inches(15,6)

x1= np.random.normal(0,2,500)
y1= np.random.normal(0,2,500)
ax[0].set_title("Original")
ax[0].scatter(x1,y1,alpha=0.5)
x2= np.random.normal(-3,0.5,100)
y2= np.random.normal(4,0.5,100)
ax[0].scatter(x2,y2,alpha=0.5)
ax[0].axis('scaled')
x3= np.random.normal(3,0.5,100)
y3= np.random.normal(4,0.5,100)
ax[0].scatter(x3,y3,alpha=0.5)
ax[0].axis('scaled')

# Show the generated data points with each group in different color.
# Using scikit-learn module,
# Run Kmeans algorithm on the data points to show three clusters in different colors.
# Run Gaussina Mixture algorithm on the data points to show three clusters in different colors.
#Kmeans algorithm
x= np.hstack([x1,x2,x3])
y= np.hstack([y1,y2,y3])
data= np.vstack([x,y]).T
kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
C=kmeans.labels_ 
ax[1].set_title("Kmeans")
ax[1].scatter(x,y,c=C,alpha=0.5)
ax[1].axis('scaled')
#Gaussian Mixture - EM Algorithm 
EM= GaussianMixture(n_components=3)
EM.fit(data)
cluster=EM.predict(data)
ax[2].set_title("GaussianMixture")
ax[2].scatter(x,y,c=cluster,alpha=0.5)
ax[2].axis('scaled')
```

```python
from sklearn import datasets
iris = datasets.load_iris()
# iris flower data set is a classic data set usually used for classification problem.
# The usual goal is to predict one of the four species using four measurements.
# The four species are ['setosa' 'versicolor' 'virginica'], represented by integer number [0,1,2,3] repectively
# Check iris.target_names and iris.target
# The fours measurements are sepal length, sepal width, petal length, petal width in centimeter - 
#   iris.data[:,0], iris.data[:,1], iris.data[:,2], iris.data[:,3] respectively
# Draw a scatter plot using 'petal length' and 'petal width'. Show 'sepal length' (iris.data[:,0]) in color
# # ADD YOUR CODE HERE
figure, ax = plt.subplots(1,1)
figure.set_size_inches(6,4)
ax.set_title('Target: Sepal Length shown in color')
ax.set_xlabel('petal length')
ax.set_ylabel('petal width')
petal_length= iris.data[:,2]
petal_width = iris.data[:,3]
sepal_length= iris.data[:,0]
ax.scatter(petal_length,petal_width,c=sepal_length)
```

```python
from sklearn.linear_model import LinearRegression
# Using scikit-learn Linear Regression,
# predict 'sepal length' using 'petal length' and 'petal width'
# Draw a scatter plot using 'petal length' and 'petal width'. Show predicted 'sepal length' in color
 # ADD YOUR CODE HERE
Regression = LinearRegression().fit(iris.data[:,-2:], sepal_length)
sepal_length_predict = Regression.predict(iris.data[:,-2:])
sepal_length = iris.data[:,0]
plt.scatter(petal_length, petal_width, c= sepal_length_predict)
plt.title("Predicted 'sepal length' in color - LinearRegression")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
# SSE = ( (t1-y1)**2 + (t2-y2)**2 + ... + (tN-yN)**2) ) / N
# ADD YOUR CODE HERE
n= len(iris.data[:,0])
SSE = np.sum((sepal_length_predict-sepal_length)**2)/n
print("Sum of Squared Errors =",SSE)
```

```python
from sklearn.tree import DecisionTreeRegressor

figure, ax = plt.subplots(1,2)
figure.set_size_inches(14,5)

# Using scikit-learn DecisionTree for regression with (min_samples_split=5)
# predict 'sepal length' using 'petal length' and 'petal width'
# Draw a scatter plot using 'petal length' and 'petal width'. Show predicted 'sepal length' in color
# Show its SSE. SSE = ( (t1-y1)**2 + (t2-y2)**2 + ... + (tN-yN)**2) ) / N
n = len(iris.data[:,0])
DTree1 = DecisionTreeRegressor(min_samples_split=5).fit(iris.data[:,-2:], sepal_length)
sepal_length_predict2 = DTree1.predict(iris.data[:,-2:])
ax[0].scatter(petal_length, petal_width, c= sepal_length_predict2)
ax[0].set_title("Predicted 'sepal length' in color - Tree(min_sample=5)")
ax[0].set_xlabel("petal length")
ax[0].set_ylabel("petal width")

SSE1 = np.sum((sepal_length_predict2 - sepal_length)**2)/n 

# Build another Decision Tree with (min_samples_split=10) and show the result again.
DTree2 = DecisionTreeRegressor(min_samples_split=10).fit(iris.data[:,-2:], sepal_length)
sepal_length_predict3 = DTree2.predict(iris.data[:,-2:])
ax[1].scatter(petal_length, petal_width, c= sepal_length_predict3)
ax[1].set_title("Predicted 'sepal length' in color - Tree(min_sample=10)")
ax[1].set_xlabel("petal length")
ax[1].set_ylabel("petal width")
plt.show()

SSE2 = np.sum((sepal_length_predict3 - sepal_length)**2)/n
print("SSE (min_sample=5) = ",SSE1)
print("SSE (min_sample=10) = ",SSE2)
```
