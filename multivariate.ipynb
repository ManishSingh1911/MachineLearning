```python
import numpy as np
import scipy.stats as ss 
import matplotlib.pyplot as plt
```

```python
data = np.random.multivariate_normal([10,20], [[9,2.5],[2.5,1]],1000)    #[width,height]
plt.scatter(data[:,0], data[:,1])
plt.axis('scaled')
ss.multivariate_normal([10,20],[[9,2.5],[2.5,1]]).pdf([(10,20),(10,25),(10,15),(5,20),(15,20)])
```
