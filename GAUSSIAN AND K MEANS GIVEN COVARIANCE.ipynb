```python
#Given covaraiance = [[25,0], [0,1]] => Sd = [5, 1] ----
# [[sd,cov][sd,cov]]
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
x1 = np.random.normal(0, 5, 100)
y1 = np.random.normal(3, 1, 100)

x2 = np.random.normal(0, 5, 100)
y2 = np.random.normal(-3, 1, 100)
figure, ax = plt.subplots(1,3)
figure.set_size_inches(13, 5)

x = np.hstack([x1, x2])
y = np.hstack([y1, y2])

data = np.vstack([x, y]).T
gmm = GaussianMixture(n_components = 2)
C = gmm.fit_predict(data)

ax[0].set_title("Original")
ax[0].scatter(x, y)

ax[1].set_title("GaussianMixture")
ax[1].scatter(data[:, 0], data[:, 1], c = C)

kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
C=kmeans.labels_ 
ax[2].set_title("Kmeans")
ax[2].scatter(x,y,c=C)
plt.show()
```
