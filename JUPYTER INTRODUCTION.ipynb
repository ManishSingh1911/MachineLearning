```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([-2,-1,2,3])
t=np.array([1,0,3,5])
plt.plot(x,t,'.')
plt.figure()
plt.scatter(x,t)
```

```python
xvals= np.arange(-5,5,0.1)
xvals
yvals= 2*xvals +3 
plt.plot(xvals,yvals)
yvals= xvals**2 
plt.plot(xvals,yvals)
yvals= np.sin(xvals)*10
plt.plot(xvals,yvals)
```

```python
x= np.array([-2,-1,2,3])
t=np.array([1,0,3,5])
plt.scatter(x,t)
import curve_fit
model=curve_fit.fit_GradientDescent(x,t)
xvals= np.arange(-3,4,0.1)
yvals= model[0]*xvals+model[1]
plt.plot(xvals,yvals)




```

```python
import sklearn.linear_model
model= sklearn.linear_model.LinearRegression()
model.fit(x.reshape(-1,1),t)
model.coef_
y= model.coef_*xvals + model.intercept_
#y=model.predict(xvals.reshape(-1,1)) can also be used 
plt.plot(xvals,y)
plt.scatter(x,t)
```

```python
x1= np.random.normal(0,1,1000)
plt.hist(x1)
plt.show()
t= np.random.normal(0,1,1000)+x1
plt.hist(t)
plt.show()
plt.scatter(x1,t,alpha=0.5)
```

```python
x1= np.random.normal(0,1,1000)
t1= np.random.normal(0,1,1000)-x1
#t1= np.random.normal(0,1,1000)+x1 for positive relationship  
plt.scatter(x1,t1,alpha=0.5)
model= sklearn.linear_model.LinearRegression()
model.fit(x1.reshape(-1,1),t1)
xvals= np.arange(-4,4,0.1)
y= model.predict(xvals.reshape(-1,1))
plt.plot(xvals,y)
```
