```python
# EM algorithm  - week 5 thursday

def GaussianMixture_k2(x):
    # given array of data points , 
    # returns multiple gaussian distribution

    #initialize 
        p_a,p_b = 0.5, 0.5 # p(a), P(b)
        mean_a, mean_b = np.random.normal(0,1,2)
        sd_a, sd_b = 1,1
        for iter in range(100):

        # E (expectation) step  is how muh is part of a or b means yellow or blue
        # b1 = scipy.stats.norm(mean_b, sd_b).pdf(x1) * p_b #P(x1|b)*P(b)
        # b2 = scipy.stats.norm(mean_b, sd_b).pdf(x2) * p_b #P(x2|b)*P(b)
        # # for a 
        # a1 = scipy.stats.norm(mean_a, sd_a).pdf(x1) * p_b #P(x1|a)*P(a)
        # a2 = scipy.stats.norm(mean_a, sd_a).pdf(x2) * p_b #P(x2|a)*P(a)

        # in one short
            b = scipy.stats.norm(mean_b, sd_b).pdf(x) * p_b
            a = scipy.stats.norm(mean_a, sd_a).pdf(x) * p_a
            t = a + b
            b = b/t
            a = a/t

        # M (Maximization)  step
            mean_b = sum(b*x) /sum(b)
            mean_a = sum(a*x) /sum(a)
            sd_b = (sum(b * (x - mean_b)**2) / sum(b))**0.5
            sd_a = (sum(a * (x - mean_a)**2) / sum(a))**0.5
            p_b = sum(b) / len(x)
            p_a = 1-p_b
        return mean_a, mean_b, sd_a, sd_b

x1 = np.array([-3,-1,-0.5,0.3,1.5,0.7,4,4.7,5,0.5,5.5])
plt.scatter(x1, x1*0)
mean_a, mean_b,sd_a,sd_b = GaussianMixture_k2(x1)

xvals = np.arange(min(x1),max(x1), 0.1)  #taking many y values not just values from x1 / the curve code
ya = scipy.stats.norm(mean_a,sd_a).pdf(xvals)
yb = scipy.stats.norm(mean_b,sd_b).pdf(xvals)
plt.plot(xvals,ya)
plt.plot(xvals,yb)
```
