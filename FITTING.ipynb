```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0, 2, 100)
t = np.random.normal(0, 4, 100) + 0.5*(x-4) **2 + 10
plt.scatter(x,t,alpha=0.5)
```

```python
import curve_fit
#model1 for line
model1= curve_fit.fit_GradientDescent(x,t)
xvals=np.arange(-5,5,0.1)
yvals= model1[0]*xvals + model1[1]
plt.scatter(x,t,alpha=0.5)
plt.plot(xvals,yvals)
plt.show()
#model2 for curve
model2= curve_fit.fit_GradientDescent2(x,t)
xvals=np.arange(-5,5,0.1)
yvals= model2[0]*xvals**2+ model2[1]*xvals + model2[2]
plt.scatter(x,t,alpha=0.5)
plt.plot(xvals,yvals)
```

```python
import sklearn.linear_model
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
#First Plot for 1 degree polynomial
poly = PolynomialFeatures(degree=1, include_bias=False)
poly_features = poly.fit_transform(x.reshape(-1, 1))
poly_regression = LinearRegression()
poly_regression.fit(poly_features,t )
poly_features_xvals = poly.fit_transform(xvals.reshape(-1, 1))
yvals = poly_regression.predict(poly_features_xvals)
plt.scatter(x,t,alpha=0.5)
plt.plot(xvals,yvals)
plt.show()
#Second plot for degree 2 polynomial (Curve)
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(x.reshape(-1, 1))
poly_regression = LinearRegression()
poly_regression.fit(poly_features,t )
poly_features_xvals = poly.fit_transform(xvals.reshape(-1, 1))
yvals = poly_regression.predict(poly_features_xvals)
plt.scatter(x,t,alpha=0.5)
plt.plot(xvals,yvals)
```
